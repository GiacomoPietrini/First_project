```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['Country'].unique()
```

```python
paesi = ['Italy', 'India', 'South Korea']

df_filtrato = df[df['Country'].isin(paesi)]
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_filtrato, x='Year', y='Total Water Consumption (Billion Cubic Meters)', hue='Country', marker='o')

plt.title('Trend del consumo dâ€™acqua nel tempo')
plt.xlabel('Anno')
plt.ylabel('Consumo totale di acqua')
plt.legend(title='Paese')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_consumo_acqua = df[['Country','Agricultural Water Use (%)','Industrial Water Use (%)','Household Water Use (%)']]
df_consumo_acqua
```

```python
statistiche = df_consumo_acqua.groupby('Country').agg({
    'Agricultural Water Use (%)': ['mean','median','std'],
    'Industrial Water Use (%)': ['mean','median','std'],
    'Household Water Use (%)': ['mean','median','std']
})

statistiche
```

```python
consumo_per_paese = df.groupby('Country')['Total Water Consumption (Billion Cubic Meters)'].sum()
consumo_per_paese
```

```python
consumo_ordinato = consumo_per_paese.sort_values(ascending= False)
consumo_ordinato
```

```python
df_2000 = df[df['Year'] == 2000]
df_2024 = df[df['Year'] == 2024]

df_crescita = df_2024[['Country','Total Water Consumption (Billion Cubic Meters)']].set_index('Country')-\
              df_2000[['Country','Total Water Consumption (Billion Cubic Meters)']].set_index('Country')

df_crescita.columns = ['Crescita assoluta']
df_top5 = df_crescita.sort_values(by = 'Crescita assoluta',ascending= False).head(5)
```
